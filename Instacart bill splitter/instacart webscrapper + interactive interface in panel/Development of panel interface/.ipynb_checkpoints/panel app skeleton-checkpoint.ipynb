Importing stuff

```python
import panel as pn
import param
import numpy as np
import pandas as pd
from tqdm.notebook import tqdm
from pdf_scarpper_functions import *
```

```python
pn.extension('mathjax','ipywidgets')
template = pn.template.BootstrapTemplate(title='Split thy bills!')
```

Inputs

```python
Input = {}
w0 = 220
Input['people'] = pn.widgets.TextInput(value='Ananth,Shubha,Tanmay,Vaasuki', name='people', width=w0)
Input['split'] = pn.widgets.Button(name='Split!', width=w0)
for item in Input.values():
    template.sidebar.append(item)
```

```python
#We have DataFrame of items. Need to be able to split it interactively
People = Input['people'].value.split(',')

checkbox_group = pn.widgets.CheckBoxGroup(
    name='', value=[], options=People,
    inline=True)
infile = 'receipt.pdf'#pdf file to be read
Items =  read_pdf(infile)#DataFrame containing the items
item_features = ['item_cost_description', 'item_quantity', 'item_name', 'item_cost']
check_boxes = []
selection_widget = {}
w1 = 500
for index, item in Items.iterrows():
    item_name = Items.iloc[index]['item_name']
    item_cost = Items.iloc[index]['item_cost']
    item_quantity = Items.iloc[index]['item_quantity']
    item_cost_description = Items.iloc[index]['item_cost_description']
    item_name_pane = pn.pane.LaTeX(object=f'[{item_cost_description}] {item_quantity} {item_name}', name='item_pane', width=w1)
    item_cost_pane = pn.pane.LaTeX(object=f'$\${item_cost}$', name='item_pane', width=50)
    selection_widget[index] = pn.widgets.CheckBoxGroup(name='', value=[], options=People, inline=True)
    template.main.append(pn.Row(item_name_pane, item_cost_pane, selection_widget[index]))
```
